## 3. Formato a mis datos

```python
datos_ma = datos.drop(['ventas_totales_en_tienda_de_cada_mes',
       'ventas_promedio_en_tienda_de_cada_mes',
       'ventas_totales_en_tienda_de_cada_sku',
       'ventas_promedio_en_tienda_de_cada_sku', 'ventas_totales_1_mes_pasado',
       'ventas_totales_2_meses_pasados', 'ventas_totales_3_meses_pasados',
       'ventas_totales_tienda_y_mes_del_mes_pasado',
       'ventas_totales_tienda_y_mes_2_pasado',
       'ventas_totales_tienda_y_mes_3_pasado',
       'ventas_promedio_tienda_y_mes_del_mes_pasado',
       'ventas_promedio_tienda_y_mes_2_pasado',
       'ventas_promedio_tienda_y_mes_3_pasado',
       'ventas_totales_tienda_y_sku_del_mes_pasado',
       'ventas_totales_tienda_y_sku_2_pasado',
       'ventas_totales_tienda_y_sku_3_pasado',
       'ventas_promedio_tienda_y_sku_del_mes_pasado',
       'ventas_promedio_tienda_y_sku_2_pasado',
       'ventas_promedio_tienda_y_sku_3_pasado'], axis=1)
```

```python
datos_ma.head()
```

## 4. Promedios Móviles

### 4.1 Pedir lo del mes pasado

```python
datos_ma['m1_pedir_mes_anterior'] =datos_ma['ventas_totales']
```

```python
datos_ma.head()
```

### 4.2 Pedir el promedio de 2 meses anteriores

```python
datos_ma['m2_pedir_2_meses_anteriores'] = datos_ma.groupby(['pdv_id', 'sku_id']).rolling(2)['ventas_totales'].mean().reset_index(drop=True)
```

```python
datos_ma.head()
```

### 4.3 Pedir el promedio de 3 meses anteriores

```python
datos_ma['m3_pedir_3_meses_anteriores'] = datos_ma.groupby(['pdv_id', 'sku_id']).rolling(3)['ventas_totales'].mean().reset_index(drop=True)
```

```python
datos_ma.head()
```

## 5. Cálculo de errores

### 5.1 Dividir por mes

```python
error_julio= datos_ma[datos_ma.mes_id==0]
error_julio.head()
```

### 5.2 Calcular erroes

```python
error_m1_julio = sqrt(mean_squared_error(error_julio['y_ventas_siguiente_mes'], error_julio['m1_pedir_mes_anterior']))
error_m2_julio = None
error_m3_julio = None
```

```python
error_m1_agosto = sqrt(mean_squared_error(error_agosto['y_ventas_siguiente_mes'], error_agosto['m1_pedir_mes_anterior']))
error_m2_agosto = sqrt(mean_squared_error(error_agosto['y_ventas_siguiente_mes'], error_agosto['m2_pedir_2_meses_anteriores']))
error_m3_agosto = None
```

```python
error_m1_septiembre = sqrt(mean_squared_error(error_septiembre['y_ventas_siguiente_mes'], error_septiembre['m1_pedir_mes_anterior']))
error_m2_septiembre = sqrt(mean_squared_error(error_septiembre['y_ventas_siguiente_mes'], error_septiembre['m2_pedir_2_meses_anteriores']))
error_m3_septiembre = sqrt(mean_squared_error(error_septiembre['y_ventas_siguiente_mes'], error_septiembre['m3_pedir_3_meses_anteriores']))
```

```python
error_m1_octubre = sqrt(mean_squared_error(error_octubre['y_ventas_siguiente_mes'], error_octubre['m1_pedir_mes_anterior']))
error_m2_octubre = sqrt(mean_squared_error(error_octubre['y_ventas_siguiente_mes'], error_octubre['m2_pedir_2_meses_anteriores']))
error_m3_octubre = sqrt(mean_squared_error(error_octubre['y_ventas_siguiente_mes'], error_octubre['m3_pedir_3_meses_anteriores']))
```

```python
error_m1_noviembre = sqrt(mean_squared_error(error_noviembre['y_ventas_siguiente_mes'], error_noviembre['m1_pedir_mes_anterior']))
error_m2_noviembre = sqrt(mean_squared_error(error_noviembre['y_ventas_siguiente_mes'], error_noviembre['m2_pedir_2_meses_anteriores']))
error_m3_noviembre = sqrt(mean_squared_error(error_noviembre['y_ventas_siguiente_mes'], error_noviembre['m3_pedir_3_meses_anteriores']))
```

```python
error_m1_diciembre = sqrt(mean_squared_error(error_diciembre['y_ventas_siguiente_mes'], error_diciembre['m1_pedir_mes_anterior']))
error_m2_diciembre = sqrt(mean_squared_error(error_diciembre['y_ventas_siguiente_mes'], error_diciembre['m2_pedir_2_meses_anteriores']))
error_m3_diciembre = sqrt(mean_squared_error(error_diciembre['y_ventas_siguiente_mes'], error_diciembre['m3_pedir_3_meses_anteriores']))
```

```python
error_m1_enero = sqrt(mean_squared_error(error_enero['y_ventas_siguiente_mes'], error_enero['m1_pedir_mes_anterior']))
error_m2_enero = sqrt(mean_squared_error(error_enero['y_ventas_siguiente_mes'], error_enero['m2_pedir_2_meses_anteriores']))
error_m3_enero = sqrt(mean_squared_error(error_enero['y_ventas_siguiente_mes'], error_enero['m3_pedir_3_meses_anteriores']))
```

```python
error_m1_febrero = sqrt(mean_squared_error(error_febrero['y_ventas_siguiente_mes'], error_febrero['m1_pedir_mes_anterior']))
error_m2_febrero = sqrt(mean_squared_error(error_febrero['y_ventas_siguiente_mes'], error_febrero['m2_pedir_2_meses_anteriores']))
error_m3_febrero = sqrt(mean_squared_error(error_febrero['y_ventas_siguiente_mes'], error_febrero['m3_pedir_3_meses_anteriores']))
```

```python
error_m1_marzo = sqrt(mean_squared_error(error_marzo['y_ventas_siguiente_mes'], error_marzo['m1_pedir_mes_anterior']))
error_m2_marzo = sqrt(mean_squared_error(error_marzo['y_ventas_siguiente_mes'], error_marzo['m2_pedir_2_meses_anteriores']))
error_m3_marzo = sqrt(mean_squared_error(error_marzo['y_ventas_siguiente_mes'], error_marzo['m3_pedir_3_meses_anteriores']))
```

## Dataframe de erroes

```python
errores_rmse = [['Julio', error_m1_julio, error_m2_julio, error_m3_julio],
               ['Agosto', error_m1_agosto, error_m2_agosto, error_m3_agosto],
               ['Septiembre', error_m1_septiembre, error_m2_septiembre, error_m3_septiembre],
               ['Octubre', error_m1_octubre, error_m2_octubre, error_m3_octubre],
               ['Noviembre', error_m1_noviembre, error_m2_noviembre, error_m3_noviembre],
               ['Diciembre', error_m1_diciembre, error_m2_diciembre, error_m3_diciembre],
               ['Enero', error_m1_enero, error_m2_enero, error_m3_enero],
               ['Febrero', error_m1_febrero, error_m2_febrero, error_m3_febrero],
               ['Marzo', error_m1_marzo, error_m2_marzo, error_m3_marzo]
              ]
rmse = pd.DataFrame(errores_rmse, columns = ['Mes', 'rmse_pedir_anterior','rmse_promedio_2_meses_anteriores','rmse_promedio_3_meses_anteriores'])
rmse
```

```python
ptl.figure(figsize=(20,6)) #impresión de la figura
ptl.plot(rmse.Mes, rmse.rmse_pedir_anterior,'-o',color="royalblue")
ptl.plot(rmse.Mes, rmse.rmse_promedio_2_meses_anteriores, '-o', color="red")
ptl.plot(rmse.Mes, rmse.rmse_promedio_3_meses_anteriores, '-o', color="darkgreen")
ptl.legend(['Pedir lo del mes anterior','Pedir promedio 2 meses anteriores','Pedir promedio 3 meses anteriores'])
ptl.ylim(1, 3.5)
ptl.ylabel('rmse')
ptl.xlabel('Mes')
ptl.title('RMSE Total en Modelos Base')
#-p une puntos con lineas
#azul menos 
```
